```python
import pandas as pd
df = pd.read_csv('crimenMeridaNOV2024.csv')
```

```python
print(df.info)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], dayfirst=True)
```

```python
print(df.info)
```

```python
df['Alerta'] = df['Alerta'].str.replace('%', '').str.replace('\xa0', '').str.strip()
df['Alerta'] = pd.to_numeric(df['Alerta']) / 100
```

```python
print(df.info)
```

```python
text_cols = ['Descripcion', 'Estado', 'Parroquia', 'Ubicación']
for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
print(df.info)
```

```python
impacto_map = {'Bajo': 1, 'Medio': 2, 'Alto': 3}
df['Impacto_Num'] = df['Impacto'].map(impacto_map)
```

```python
print(df.info)
```

```python
df['Dia_Semana'] = df['Fecha'].dt.day_name()
```

```python
df = df.dropna(subset=['Codigo'])
```

```python
df.to_csv('crimen_merida_depurado.csv', index=False)
```
